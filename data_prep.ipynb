```python
import pandas as pd
```

```python
df = pd.read_csv("pennis.csv")
alco = pd.read_csv("alcohol_consumption.csv")
```

```python
alco = alco.rename(columns = {"Unnamed: 0": "country", "Unnamed: 2": "Type", 
                              "Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol).1": "Alc 2015"})
```

```python
alco = alco.iloc[1:,[0,2,4]]
```

```python
alco = alco[alco["Type"] == "All types"]
```

```python
ctry = list(df.country.unique())
```

```python
country = ["Austria", "Belgium", "Switzerland",
                  "Czechia", "Germany", "Estonia", 
                  "Spain", "Finland", "France", 
                  "United Kingdom of Great Britain and Northern Ireland",
                  "Hungary", "Ireland", "Israel", "Iceland", 
                  "Italy", "Lithuania", "Netherlands", 
                  "Norway", "Poland", "Portugal", 
                  "Russian Federation", "Sweden", "Slovenia"]
```

```python
mob = {}
for i in range(0, len(country)):
    mob[ctry[i]] = country[i] 
    
```

```python
df = df.iloc[:,1:]
```

```python
df.country.replace(mob, inplace= True)
```

```python
df1 = pd.merge(df,alco, on="country")
```

```python
import numpy as np
```

```python
rand = []
for i in list(df.country.unique()):
    rand.append(list(np.random.normal(df1[df1['country']==i]['Alc 2015'],1)))
```

```python
randd = [x for y in rand for x in y]
```

```python
fin = []
for i in randd:
    fin.append(int(round(i,0)))
```

```python
df1["consumption"] = fin
```

```python
final = df1.drop(["country", "Alc 2015", "gender"], axis = 1)
```

```python
final.to_csv("modified.csv", index=False)
```
